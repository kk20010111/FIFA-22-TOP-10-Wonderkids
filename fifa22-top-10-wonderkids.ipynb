```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, RegularPolygon
from matplotlib.path import Path
from matplotlib.projections.polar import PolarAxes
from matplotlib.projections import register_projection
from matplotlib.spines import Spine
from matplotlib.transforms import Affine2D

def radar_factory(num_vars, frame='circle'):
    """Create a radar chart with `num_vars` axes."""
    theta = np.linspace(0, 2*np.pi, num_vars, endpoint=False)

    class RadarAxes(PolarAxes):
        name = 'radar'
        def __init__(self, *args, **kwargs):
            super().__init__(*args, **kwargs)
            self.set_theta_zero_location('N')

        def fill(self, *args, closed=True, **kwargs):
            """Override fill so that line is closed by default."""
            return super().fill(closed=closed, *args, **kwargs)

        def plot(self, *args, **kwargs):
            """Override plot so that line is closed by default."""
            lines = super().plot(*args, **kwargs)
            for line in lines:
                self._close_line(line)

        def _close_line(self, line):
            x, y = line.get_data()
            if x[0] != x[-1]:
                x = np.concatenate((x, [x[0]]))
                y = np.concatenate((y, [y[0]]))
                line.set_data(x, y)

        def set_varlabels(self, labels):
            self.set_thetagrids(np.degrees(theta), labels)

        def _gen_axes_patch(self):
            if frame == 'circle': return Circle((0.5, 0.5), 0.5)
            elif frame == 'polygon': return RegularPolygon((0.5, 0.5), num_vars, radius=0.5, edgecolor="k")
            else: raise ValueError("Unknown value for 'frame': %s" % frame)

        def draw(self, renderer):
            if frame == 'polygon':
                self.grid(True)
                r_grids = self.yaxis.get_gridlines()
                for gr in r_grids: gr.set_color('#333333')
            super().draw(renderer)

        def _gen_axes_spines(self):
            if frame == 'circle': return super()._gen_axes_spines()
            elif frame == 'polygon':
                spine = Spine(axes=self, spine_type='circle', path=Path.unit_regular_polygon(num_vars))
                spine.set_transform(Affine2D().scale(.5).translate(.5, .5) + self.transAxes)
                return {'polar': spine}
            else: raise ValueError("Unknown value for 'frame': %s" % frame)

    register_projection(RadarAxes)
    return theta

file_path = ""
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename == 'players_22.csv':
            file_path = os.path.join(dirname, filename)
            break
    if file_path: break

try:
    if not file_path: raise FileNotFoundError
    df = pd.read_csv(file_path, low_memory=False)

    pedri_df = df[df['short_name'] == 'Pedri']
    if pedri_df.empty: raise ValueError("Pedri not found in this specific dataset.")
    
    player_info = pedri_df.iloc[0]
    
    categories = ['PACE', 'SHOOTING', 'PASSING', 'DRIBBLING', 'DEFENDING', 'PHYSICALITY']
    values = [player_info['pace'], player_info['shooting'], player_info['passing'], 
              player_info['dribbling'], player_info['defending'], player_info['physic']]
    
    # Radar Styling (Modern dark interface)
    fig, ax = plt.figure(figsize=(10, 10), facecolor='#111111').add_subplot(111, projection='radar')
    theta = radar_factory(len(categories), frame='polygon')
    ax.set_facecolor('#111111')
    
    # Custom colors: Barcelona Cyan and Gold
    ax.plot(theta, values, color='#00FFFF', marker='o', linewidth=3)
    ax.fill(theta, values, facecolor='#00FFFF', alpha=0.3)
    
    ax.set_varlabels(categories)
    ax.set_rgrids([0, 20, 40, 60, 80, 100], labels=[], color='#555555')
    ax.set_ylim(0, 100)
    
    # Titles and Labels (English scouting style)
    plt.title(f"PEDRI - PLAYER PROFILE (FIFA 22)", size=22, color='white', y=1.05)
    
    label_info = f"{player_info['club_name'].upper()} | AGE: {player_info['age']} | OVR: {player_info['overall']} | POT: {player_info['potential']}"
    fig.text(0.5, 0.93, label_info, horizontalalignment='center', color='#AAAAAA', size=14)
    
    grid_lines = ax.xaxis.get_gridlines()
    for line in grid_lines: line.set_color('#333333')
    
    for label, angle in zip(ax.get_xticklabels(), theta):
        label.set_color('white')
        label.set_fontsize(12)
        
    print(f"📊 Analyzing data for Pedri (OVR {player_info['overall']} / POT {player_info['potential']})")
    print(f"Stats used: {list(zip(categories, values))}")
    
    # Simulate data rendering
    plt.show()

except FileNotFoundError:
    print("⚠️ 'players_22.csv' not found. Please re-check the input panel.")
except ValueError as e:
    print(f"⚠️ Data Error: {e}")
except Exception as e:
    print(f"⚠️ Graphing Error: {e}")
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = ""
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename == 'players_22.csv':
            file_path = os.path.join(dirname, filename)
            break
    if file_path: break

try:
    if not file_path: raise FileNotFoundError
    df = pd.read_csv(file_path, low_memory=False)

    player1_name = 'Pedri'
    player2_name = 'J. Bellingham'

    p1_df = df[df['short_name'] == player1_name]
    if p1_df.empty: raise ValueError(f"{player1_name} not found.")
    p1_info = p1_df.iloc[0]

    p2_df = df[df['short_name'] == player2_name]
    if p2_df.empty: raise ValueError(f"{player2_name} not found.")
    p2_info = p2_df.iloc[0]

    categories = ['PACE', 'SHOOTING', 'PASSING', 'DRIBBLING', 'DEFENDING', 'PHYSICALITY']
    categories_plot = [*categories, categories[0]]

    v1 = [p1_info['pace'], p1_info['shooting'], p1_info['passing'], 
          p1_info['dribbling'], p1_info['defending'], p1_info['physic']]
    v2 = [p2_info['pace'], p2_info['shooting'], p2_info['passing'], 
          p2_info['dribbling'], p2_info['defending'], p2_info['physic']]
    
    v1_plot = [*v1, v1[0]]
    v2_plot = [*v2, v2[0]]

    angles = np.linspace(0, 2 * np.pi, len(categories_plot))

    fig, ax = plt.subplots(figsize=(9, 9), subplot_kw=dict(polar=True), facecolor='#111111')
    ax.set_facecolor('#111111')

    # Plot Player 1 (Cyan - Pedri)
    ax.plot(angles, v1_plot, color='#00FFFF', linewidth=3, label=f"{player1_name} (OVR {p1_info['overall']})")
    ax.fill(angles, v1_plot, color='#00FFFF', alpha=0.3)

    # Plot Player 2 (Gold - Bellingham)
    ax.plot(angles, v2_plot, color='#FFD700', linewidth=3, label=f"{player2_name} (OVR {p2_info['overall']})")
    ax.fill(angles, v2_plot, color='#FFD700', alpha=0.3)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, color='white', size=13, weight='bold')

    ax.set_yticks([20, 40, 60, 80, 100])
    ax.set_yticklabels(['20', '40', '60', '80', '100'], color='#777777')
    ax.set_ylim(0, 100)

    ax.spines['polar'].set_color('#444444')
    ax.grid(color='#333333', linestyle='--')

    plt.title(f"Wonderkid Comparison: {player1_name} vs. {player2_name}", 
              size=20, color='white', y=1.1, weight='bold')
    legend = ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), facecolor='#222222', edgecolor='#444444')
    plt.setp(legend.get_texts(), color='white')

    print(f"✅ Successfully compared {player1_name} and {player2_name}.")
    plt.show()

except Exception as e:
    print(f"⚠️ Error during comparison: {e}")
```
